```python
import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]
z = [200, 25, 400, 3330, 100]

Titulo = "Scatterplot: Gráfico de Disperção"
Eixo_x = "Eixo X"
Eixo_y = "Eixo Y"

# Legendas
plt.title(Titulo)
plt.xlabel(Eixo_x)
plt.ylabel(Eixo_y)

# Referencias da Aula
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

# . scatter
plt.scatter(x, y, label = "Meus Pontos", color = "Red", marker = ".", s = z)
plt.plot(x, y, color = "Green", linestyle = "--")

# Legenda
plt.legend()

# Salvar
plt.savefig("Figura.png", dpi = 300) # dpi é o tamanho da figura em png
plt.savefig("Figura.pdf")
```
