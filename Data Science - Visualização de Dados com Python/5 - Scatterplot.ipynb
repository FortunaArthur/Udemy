```python
import matplotlib.pyplot as plt

x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]

Titulo = "Scatterplot: Gráfico de Disperção"
Eixo_x = "Eixo X"
Eixo_y = "Eixo Y"

# Legendas
plt.title(Titulo)
plt.xlabel(Eixo_x)
plt.ylabel(Eixo_y)

# . scatter
plt.scatter(x, y, label = "Meus Pontos", color = "Red")
plt.plot(x, y)

# Legenda
plt.legend()
```
